```python
import numpy as np
import matplotlib.pyplot as plt
from utls import algorithms,simulations,plots
```

```python
# options
dt = 0.001
timesteps = 100000
delays = 128

# simulate system
# lorenz, duffing_oscillator, r√∂ssler, vanderpol_oscillator
time_series = simulations.simulate_system('vanderpol_oscillator',dt=dt, timesteps=timesteps)
t_train = time_series['t_train']
X_train = time_series['X_train']
t_test = time_series['t_test']
X_test = time_series['X_test']
```

```python
# run algorithm
dmd = algorithms.DMD()
dmd.fit(X_train, dt, delays=delays, trunc_mode='rank', s_thresh=24)
X_pred = dmd.reconstruct(t_test)
Vh = dmd.Vh_
```
